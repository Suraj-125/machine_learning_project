<h1 style = "background-color: #000; color : #fff; padding:10px 5px">Predicting the House Pricing using Simple linear Regression</h1>

```python
# Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

# Using inline magic
%matplotlib inline
```

```python
# Reading the csv file and storing into DataFrame
df = pd.read_csv("homeprices.csv")
df
```

```python
# Visulaizing the given data using matplotlib

plt.title("Housing Price Prediction")
plt.xlabel("Area")
plt.ylabel("Prices")
plt.scatter(df.area, df.price, marker = "+", color= "purple")
```

```python
# Initializing the linear_model
reg = linear_model.LinearRegression()
# Fitting x as area and y as price for prediction
reg.fit( df[["area"]], df.price)
```

```python
# Predicting the price
reg.predict(np.array([[3540]]))
```

```python
# Calculating the coefficient
reg.coef_
```

```python
# Calculating the intercept
reg.intercept_
```

```python
# Visulaizing the plot using the prediction model
plt.title ("Housing Price Prediction")
plt.xlabel("Area")
plt.ylabel("Prices")
plt.scatter(df.area, df.price, marker = "*", color = "red")
# Ploting the plot using prediction model
plt.plot(df.area, reg.predict(df[["area"]]), color = "purple")
plt.show()
```

```python
# Raeding another csv file to predict the housing price
d = pd.read_csv("areas.csv")
d.head()
```

```python
# Storing predicted price into a variable
prediction = reg.predict(d)
prediction
```

```python
# Storing the predicted housing price into a price column
d["prices"] = prediction
```

```python
# writing the predicted price into another csv file
d.to_csv("prediction.csv", index = False)
```
